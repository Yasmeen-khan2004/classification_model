```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

## Decision Tree implementation

```python
# for classificatio 
from sklearn.tree import DecisionTreeClassifier
# for regression
# from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(x_train , y_train)    # training our model
```

```python
# plot_tree(dtc)   -> used for show all bhind the seen tree
```

```python
print(dtc.score(x_train , y_train))
print(dtc.score(x_test,y_test))

# training score-> high
# testing score -> low   
# overfit


# pruring process -- cut the branches of decision tree   --> pre pruring , post pruning
```
